```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in ['Inventory Level','Units Sold','Units Ordered','Price','Discount','Competitor Pricing','Demand']:
    sns.histplot(data[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
for col in ['Inventory Level','Units Sold','Units Ordered','Price','Discount','Competitor Pricing','Demand']:
    sns.boxplot(x=data[col])
    plt.title(col)
    plt.show()
```

```python
data['Category'].value_counts()
data['Weather Condition'].value_counts()
data['Seasonality'].value_counts()
```

```python
sns.countplot(x=data['Category'])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x=data['Price'], y=data['Demand'])
sns.scatterplot(x=data['Inventory Level'], y=data['Demand'])
sns.scatterplot(x=data['Competitor Pricing'], y=data['Demand'])
sns.scatterplot(x=data['Units Sold'], y=data['Demand'])
```

```python
sns.boxplot(x='Seasonality', y='Demand', data=data)
sns.boxplot(x='Weather Condition', y='Demand', data=data)
sns.boxplot(x='Promotion', y='Demand', data=data)
sns.boxplot(x='Category', y='Demand', data=data)
```

```python
pd.crosstab(data['Region'], data['Seasonality'])
sns.heatmap(pd.crosstab(data['Region'], data['Seasonality']), annot=True)
```
